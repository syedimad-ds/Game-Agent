```python
# Cell 3: Verification & Random Agent Demo (Fixed)
import gymnasium as gym
import base64
from IPython.display import HTML
from IPython.display import display as ipy_display
from gymnasium.wrappers import RecordVideo

# 1. Environment Creation
env = gym.make("LunarLander-v3", continuous=True, render_mode="rgb_array")

# 2. Video recording
env = RecordVideo(env, video_folder='./videos', name_prefix='random_agent')

# 3. Game Start (1 Episode)
observation, info = env.reset()
done = False

print("🚀 Starting Random Agent Episode...")

while not done:
    random_action = env.action_space.sample()
    observation, reward, terminated, truncated, info = env.step(random_action)
    done = terminated or truncated

env.close()
print("💥 Episode Finished (Usually a crash for random agent).")

# 4. Function to Display the Video
def show_video(video_path):
    video = open(video_path, 'rb').read()
    encoded = base64.b64encode(video)
    ipy_display(HTML(data='''<video alt="test" autoplay
                loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii'))))

# Display the recorded video
show_video('./videos/random_agent-episode-0.mp4')
```

```python
# Cell 4: Training the Lunar Lander with DQN (Fixed for Discrete Actions)
import gymnasium as gym
from stable_baselines3 import DQN
from stable_baselines3.common.evaluation import evaluate_policy
import os

# 1. Environment Set Up (FIXED: Removed continuous=True)
env = gym.make("LunarLander-v3")

print("🧠 Setting up the AI Brain (DQN Model) for Discrete Actions...")

# 2. Model Parameters Setup
model = DQN(
    "MlpPolicy",
    env,
    learning_rate=0.0005,
    buffer_size=50000,
    learning_starts=1000,
    batch_size=64,
    gamma=0.99,
    exploration_fraction=0.1,
    exploration_initial_eps=1.0,
    exploration_final_eps=0.02,
    target_update_interval=250,
    train_freq=4,
    verbose=1,
    tensorboard_log="./lunar_lander_tensorboard/"
)

print("🚀 Starting Training... (This will take 5-10 minutes)")

# 3. Start Training (100k steps is a good starting point)
model.learn(total_timesteps=100000, progress_bar=True)

print("✅ Training Completed!")

# 4. Save the Model
model_path = "dqn_lunar_lander"
model.save(model_path)
print(f"💾 Model saved as {model_path}.zip")

# 5. Evaluate the Model (Testing)
print("🧪 Evaluating the trained model...")
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10, render=False)
print(f"📊 Mean Reward: {mean_reward:.2f} +/- {std_reward:.2f}")

env.close()
```

```python
# Cell 6: Resume Training to create a "Pro Agent"
import gymnasium as gym
from stable_baselines3 import DQN
from stable_baselines3.common.evaluation import evaluate_policy
import os

env = gym.make("LunarLander-v3")

print("🏋️ Loading the existing model to continue training...")
# Load the model we just trained
model = DQN.load("dqn_lunar_lander", env=env)

# Optionally, you can change the exploration rate so it doesn't do too many random things now
model.exploration_final_eps = 0.01

print("🚀 Starting Extended Training... (This might take 15-20 minutes)")
# Let's train it for another 400,000 steps to reach 500k total
model.learn(total_timesteps=400000, progress_bar=True, reset_num_timesteps=False)

print("✅ Extended Training Completed!")

# Save the super-trained model
pro_model_path = "dqn_lunar_lander_pro"
model.save(pro_model_path)
print(f"💾 Pro Model saved as {pro_model_path}.zip")

# Evaluate the pro model
print("🧪 Evaluating the PRO model...")
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10, render=False)
print(f"📊 PRO Mean Reward: {mean_reward:.2f} +/- {std_reward:.2f}")

env.close()
```

```python
# Cell 7: Record the Best "PRO" Landing
import gymnasium as gym
from stable_baselines3 import DQN
from gymnasium.wrappers import RecordVideo
import base64
from IPython.display import HTML
from IPython.display import display as ipy_display
import shutil
import os

print("🎬 Setting up the camera for the highlight reel...")

# Clean up old videos so we only get the new ones
if os.path.exists('./pro_videos'):
    shutil.rmtree('./pro_videos')

# 1. Environment Setup (Record every episode)
env = gym.make("LunarLander-v3", render_mode="rgb_array")
# We record every episode (episode_trigger = lambda x: True)
env = RecordVideo(env, video_folder='./pro_videos', name_prefix='pro_agent', episode_trigger=lambda x: True)

# 2. Load PRO Model
model = DQN.load("dqn_lunar_lander_pro")

# 3. Play 5 episodes and track the best one
best_reward = -float('inf')
best_episode = 0

for episode in range(5):
    obs, info = env.reset()
    done = False
    total_reward = 0

    while not done:
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward
        done = terminated or truncated

    print(f"Episode {episode}: Score = {total_reward:.2f}")
    if total_reward > best_reward:
        best_reward = total_reward
        best_episode = episode

env.close()
print(f"🌟 Best Landing was Episode {best_episode} with Score {best_reward:.2f}")

# 4. Display the BEST video
def show_best_video(episode_num):
    video_path = f'./pro_videos/pro_agent-episode-{episode_num}.mp4'
    video = open(video_path, 'rb').read()
    encoded = base64.b64encode(video)
    ipy_display(HTML(data='''<video alt="test" autoplay
                loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii'))))

print("🎥 Showing the Best Landing Video:")
show_best_video(best_episode)

# Rename the best video so it's easy to download later
os.rename(f'./pro_videos/pro_agent-episode-{best_episode}.mp4', './best_landing.mp4')
print("✅ Best video saved as 'best_landing.mp4'. Please download it from the left panel!")
```
